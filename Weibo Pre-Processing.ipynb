# Reading from CSV

```python
import csv
```

```python
def csv_to_texts(filename):
    texts = []
    with open(filename, 'r') as f:
        reader = csv.DictReader(f)
        for row in reader:
            texts.append(row['text'])
    return texts
```

```python
texts_by_week = [[]] # placeholder for more intuitive indexing
for i in range(1, 9):
    filename = f'week{i}_clean.csv'
    texts_by_week.append(csv_to_texts(filename))
```

```python
for texts in texts_by_week:
    print(texts[:2])
```

# Data Cleaning

```python
import string
```

```python
english_chars = set(string.printable)
```

```python
from stopwordsiso import stopwords
```

```python
stopwords_zh = set(stopwords(["zh"]))
```

```python
#a list of Chinese stopwords
stopwords_zh
```

```python
#a list of punctuations
```

```python
chinese_stop_chars = {'！', '：', '，', '。', '？', '【', '】', '《', '》', '；'}
```

```python
stop_chars = english_chars.union(chinese_stop_chars).union(stopwords_zh)
```

```python
#detection of emojis
```

```python
!pip install --upgrade harvesttext
```

```python
from harvesttext import HarvestText
```

```python
ht = HarvestText()
```

```python
def clean_text(text):
    ht_text = ht.clean_text(text)
    return ''.join(c for c in ht_text if c not in stop_chars)
```

```python
clean_texts_by_week = [[clean_text(text) for text in texts] for texts in texts_by_week]
```

```python
clean_texts_by_week.append([clean_text(text) for text in texts_by_week[8]])
```

# Tokenize texts

```python
import zh_core_web_md
nlp = zh_core_web_md.load()
```

```python
len(clean_texts_by_week)
```

```python
tokenized_texts_by_week = []
for week, clean_texts in enumerate(clean_texts_by_week):
    if week == 0:
        continue
    tokenized_texts = nlp.tokenizer.pipe(clean_texts)
    all_tokens = []
    #print out to keep track of the progress
    for i, text in enumerate(tokenized_texts):
        for token in text:
            all_tokens.append(str(token))
        if i % 10000 == 0:
            print(week, i)
    with open(f'week{week}_tokenized.txt', 'w') as f:
        f.write(' '.join(all_tokens))
```
